```python
Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class
```

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head(3)
```

```python
def class_(rating):
    if rating <= 2:
        return 'низкий рейтинг'
    elif rating <= 4:
        return 'средний рейтинг'
    elif rating >= 4.5 and rating <= 5.0:
        return 'высокий рейтинг'
    
```

```python
ratings['class'] = ratings['rating'].apply(class_)
ratings[['movieId', 'rating', 'class']].sample(6)
```

```python
Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который КАЖДОЙ СТРОКЕ сможет выставить
географическую принадлежность определенному региону. 
Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. 
Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
```

```python
df = pd.read_csv('keywords.csv')
df.head()
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
# функция: если для города есть регион, то пишем его в столбец region 
def geo_classifier(keyword):
    for region, cities_list in geo_data.items():
        for citi in cities_list:
            if citi in keyword:
                return region 
    return 'undefined'
            
```

```python
df['region'] = df['keyword'].apply(geo_classifier)
df.sample(15)
```

```python
years= list(range(1950, 2011))
```

```python
def production_year(row):
    for year in years:
        if str(year) in row:
            return year
    return 1900
```

```python
movies['release'] = movies['title'].apply(production_year)
movies.head(5) # применяем функцию к столбцу (или здесь лучше применить ко всему df!?)
```

```python
merged = movies.merge(ratings, how = 'left', on = 'movieId') 
merged.groupby('title').mean('release').sort_values('rating', ascending=False).head(5) 
# группируем по названию фильма, находим средний рейтинг для каждого года, затем сортируем по рез-т (рейтинг) по убыванию
```
